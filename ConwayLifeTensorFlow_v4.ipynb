```
from __future__ import absolute_import, division, print_function, unicode_literals
```

```
import tensorflow as tf
```

```
import pathlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#np.set_printoptions(precision=4)
```

```
data_set_size = tf.constant([10_000],dtype=tf.int32)
```

```
life_density_kernal = tf.constant([1,1,1,1,10,1,1,1,1],dtype=tf.float32,name='life_density')
def alive(x):
  x1 = tf.multiply(x,life_density_kernal)
  x_sum = tf.reduce_sum(x1)
  return tf.cast(tf.where(((x_sum == 3)|(x_sum == 12)|(x_sum == 13)),1,0),dtype=tf.float32).numpy()
```

```
def get_data_examples(count=data_set_size):
  for i in range(count[0]):
    data_element = tf.round(tf.random.uniform([9],minval=0,maxval=1,dtype=tf.float32),name="data")
    label = alive(data_element)
    yield data_element[:],label
```

```
def get_data_set():
  data_gen = tf.data.Dataset.from_generator(get_data_examples,(tf.float32,tf.float32))
```

```
# %time get_data_set()
```

```
from tensorflow.keras import models
from tensorflow.keras import layers
```

```
model1 = models.Sequential()
model1.add(layers.Dense(3, activation='relu', input_shape=(9,)))
model1.add(layers.Dense(1, activation='sigmoid'))
model1.compile(optimizer=tf.keras.optimizers.Adam(0.001),
              loss='mse',       # mean squared error
              metrics=['accuracy'])
```

```
model2 = models.Sequential()
model2.add(layers.Dense(3, activation='relu', input_shape=(9,)))
model2.add(layers.Dense(1, activation='sigmoid'))
model2.compile(optimizer=tf.keras.optimizers.Adam(0.003),
              loss='mse',       # mean squared error
              metrics=['accuracy'])
```

```
model3 = models.Sequential()
model3.add(layers.Dense(3, activation='relu', input_shape=(9,)))
model3.add(layers.Dense(1, activation='sigmoid'))
model3.compile(optimizer=tf.keras.optimizers.Adam(0.005),
              loss='mse',       # mean squared error
              metrics=['accuracy'])
```

```
dataset = tf.data.Dataset.from_generator(get_data_examples,(tf.float32,tf.float32),output_shapes=((9,),())) 
```

```
dataset1 = dataset.batch(1)
%time model1.fit(dataset1,epochs=10)
```

```
dataset2 = dataset.batch(1)
%time model2.fit(dataset2,epochs=10)
```

```
dataset3 = dataset.batch(1)
%time model3.fit(dataset3,epochs=10)
```

```
test_set_size = tf.constant([100_000],dtype=tf.int32)

def get_test_examples(count=test_set_size):
  for i in range(count[0]):
    data_element = tf.round(tf.random.uniform([9],minval=0,maxval=1,dtype=tf.float32),name="data")
    label = alive(data_element)
    yield data_element[:],label

testset = tf.data.Dataset.from_generator(get_test_examples,(tf.float32,tf.float32),output_shapes=((9,),()))
```

```
testset1 = testset.batch(1000)
model1.evaluate(testset1)
```

```
testset2 = testset.batch(1000)
model2.evaluate(testset2)
```

```
testset3 = testset.batch(1000)
model3.evaluate(testset3)
```

```
testset_one= testset.batch(1)
```

```
counter = 0
for i in testset_one:
  counter += 1
  t = i
  #print(t[1])
  predicted = model1.predict(t)
  result = tf.reshape(tf.round(predicted),(1,))
  #tf.print(t,result,summarize=-1)
  equals = tf.equal(result,t[1])
  #print(equals)
  if tf.reduce_all(equals) == False:
    print("failed", t, predicted, result)
  if counter%1000 == 0: print("counting", counter)
print("done",counter)
```
